### Data to create prompts

```python
name_of_competition=["world Volleyball championship","World Basketball championship","world Football championship", "Australian Open"]
score=["3-2","56-73","2-0","2 set to 1"]
nat1=["Serbia","Australia","USA","USA"]
nat2=["Italy","Usa","Netherlands","Spain"]
date=["2018","2018","2019","2020"]
state=["Japan","Australia","France","Australia"]
trainer_name=["Zoran TerziÄ‡","Dawn Staley","Jill Ellis","Rick Macci"]
winner=["Serbia","Usa","Usa","Sofia Kenin"]
list_unroll_ans=[]
list_unroll_prompts=[]
list_unroll_right_ans=[]
list_unroll_resp = []

MODEL = 'llama3'

for z in range(4):

  chat_=[]

  list_of_prompt=["Who won the " + name_of_competition[z] +"?",
                "Who won the " + name_of_competition[z] +" with a score of "+score[z]+"?",
                "Who won the " + name_of_competition[z] +" where the final was between "+nat1[z]+" and "+nat2[z]+"?",
                "Who won the " + name_of_competition[z] +" on the year "+date[z]+"?",
                "Who won the " + name_of_competition[z] +" played in "+state[z]+"?",
                "Who won the " + name_of_competition[z] +" with the trainer "+trainer_name[z]+"?"]

  for j in list_of_prompt:

    chat_.append(
      {'role': 'user', 'content': j}
    )
    prompt_question=j

    prompt=prompt_question + '\n'+  prompt_response_format

    model_answer = ollama.chat(model=MODEL, messages=chat_)

    chat_.append(model_answer['message']['content'])

    list_unroll_resp.append(model_answer['message']['content'])
    list_unroll_ans.append("Prova")
    list_unroll_prompts.append(prompt_question)
    list_unroll_right_ans.append(winner[z])
```

```python
#Create and update df with answers
df=pd.DataFrame()
df['Answer_model']=list_unroll_ans
df['Question_model']=list_unroll_prompts
df['Answer_real']=list_unroll_right_ans
```
